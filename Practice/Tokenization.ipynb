TOKENISATION EXAMPLE

```python
corpus = """Hello Welcome, to Krish's NLP tutorials. 
Watch the entire tutorial! to become an expert in NLP"""
```

```python
corpus
```

```python
print (corpus)
# /n goes away
```

```python
#Tokenisation
# sentence to paragraph 

from nltk.tokenize import sent_tokenize
```

```python
sent_tokenize(corpus)
```

```python
#save this in document
documents = sent_tokenize(corpus)
```

```python
type(documents)
```

```python
for sentence in documents:
    print(sentence)
```

```python
# tokenisation 
# convert a paragraph into words, convert sentence into words
from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
#all characters like comma, fullstop is also splitted
# doesnot split the appostrophe thing
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
#tokenises the appostrphe and all also
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
#fullstop is not treated as a seperate word here. But wrt the last word, fullstop will be 
#seperated.
```
